```python
import pandas as pd

df_customer_signups = pd.read_csv('customer_signups.csv')
df_customer_signups
```

```python
df_customer_signups.info()
```

```python
print(df_customer_signups['signup_date'].unique())
```

```python
df_customer_signups['signup_date'] = pd.to_datetime(df_customer_signups['signup_date'], format='%d-%m-%y', errors='coerce')
df_customer_signups
```

```python
print(df_customer_signups['plan_selected'].unique())
```

```python
df_customer_signups['plan_selected'] = df_customer_signups['plan_selected'].str.lower()
df_customer_signups['plan_selected'] = df_customer_signups['plan_selected'].replace('unknownplan', pd.NA)
df_customer_signups['plan_selected'] = df_customer_signups['plan_selected'].replace({'basic':'Basic','premium': 'Premium', 'prem':'Premium','pro': 'Pro'})
df_customer_signups
```

```python
print(df_customer_signups['gender'].unique())
```

```python
df_customer_signups['gender'] = df_customer_signups['gender'].str.lower()
df_customer_signups['gender'] = df_customer_signups['gender'].replace('123', pd.NA)
df_customer_signups['gender'] = df_customer_signups['gender'].replace({'female':'Female','male': 'Male','non-binary': 'Non-Binary', 'other': 'Other'})
df_customer_signups
```

```python
print("Duplicates removed:", df_customer_signups['customer_id'].duplicated().sum())
df_customer_signups = df_customer_signups.drop_duplicates(subset='customer_id')
df_customer_signups
```

```python
df_customer_signups = df_customer_signups.dropna(subset=['email', 'region', 'age'])
df_customer_signups
```
